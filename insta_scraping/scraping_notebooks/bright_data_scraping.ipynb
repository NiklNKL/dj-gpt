Formatting response

```python
import pandas as pd
from pathlib import Path

root_path = Path(".").resolve()

comment_df = pd.read_csv(f"{root_path}/scraped_data/dorfterror_comments.csv")
display(comment_df.head())
```

```python
comment_df = comment_df[['comment_user', 'comment_user_url', 'comment_date', 'comment',
       'likes_number', 'replies_number', 'replies', 'hashtag_comment',
       'tagged_users_in_comment', 'post_url', 'comment_id', 'post_id',
       'timestamp']]
```

```python
from dorfterror_database.formatting_utils import format_list
temporary_df = comment_df.dropna(subset=["replies"]).reset_index().copy()
all_reply_comments = []
for _, row in temporary_df.iterrows():
    replies = format_list(row['replies'])
    for i, reply in enumerate(replies):
        new_comment = {
            'comment_id': int(str(row['comment_id'])+str(i)),
            'parent_comment_id': row['comment_id'],
            'post_id': row['post_id'],
            'post_url': row['post_url'],
            'comment_user': reply.get("reply_user"),
            'comment_user_url': f"https://www.instagram.com/{reply.get('reply_user')}/",
            'replies_number': None,
            'replies': None,
            'comment': reply.get('reply'),
            'likes_number': reply.get('likes_number'),
            'comment_date': reply.get('reply_date'),
            'hashtag_comment': reply.get('reply'),
            'tagged_users_in_comment': reply.get('reply'),
            'timestamp': row['timestamp'],
        }
        all_reply_comments.append(new_comment)
comment_df = pd.concat([comment_df, pd.DataFrame(all_reply_comments)], ignore_index=True)
print(comment_df.shape)
display(comment_df.head())
```

```python
comment_df.to_parquet(f"{root_path}/filtered_datasets/all_dorfterror_comments.parquet", index=False)
```

## Get Missing Profiles

Missing Collaborators

```python
from sqlalchemy.orm import sessionmaker

from dorfterror_database.database_structure import Base, Post, Profile
import json
from sqlalchemy import create_engine
from pathlib import Path

root_path = Path(".").resolve()
print(root_path)
engine = create_engine(f"sqlite:///{root_path}/dorfterror_database/dorfterror.db")
Base.metadata.create_all(engine)

Session = sessionmaker(bind=engine)
session = Session()

posts = session.query(Post).all()
all_collaborators = set()

for post in posts:
    if post.user_posted and post.user_posted != "dorfterror":
        all_collaborators.add(post.user_posted)
    
    if post.coauthor_producers:
        try:
            coauthors = post.coauthor_producers
            if isinstance(coauthors, str):
                coauthors = json.loads(coauthors)  # falls als String gespeichert
            all_collaborators.update(coauthors)
        except Exception as e:
            print(f"Error while parsing coauthor: {e}")

existing_profiles = set([
    profile.username for profile in session.query(Profile.username).all()
])

missing_collaborators = all_collaborators - existing_profiles

print("Fehlende Profile:")
print(sorted(missing_collaborators))
```

### Scraping more posts from related artists to get more data

```python
from datetime import datetime
from utils import add_project_root_to_path
add_project_root_to_path()
from dorfterror_database.database_structure import Post, Collaborator, Profile
from collections import Counter
from sqlalchemy.orm import sessionmaker
from sqlalchemy import create_engine
import pandas as pd

engine = create_engine(f"sqlite:///./dorfterror_database/dorfterror.db")
Session = sessionmaker(bind=engine)
session = Session()

# Query posts from 2022 or newer
recent_posts = session.query(Post).filter(Post.date_local >= datetime(2022, 1, 1)).all()
recent_collaborators = session.query(Collaborator).filter(Collaborator.post_id.in_([post.id for post in recent_posts])).all()
profile_post_counts = session.query(Profile).filter(Profile.id.in_([c.profile_id for c in recent_collaborators])).all()


collaborator_names = [c.username for c in recent_collaborators]
collaborator_counts = Counter(collaborator_names)

print("Collaborators sorted by frequency:")
for name, count in collaborator_counts.most_common()            :

    table_data = []
    for name, count in collaborator_counts.most_common():
        post_count = next((p.post_count for p in profile_post_counts if p.username == name), None)
        profile_category = next((p.category_name for p in profile_post_counts if p.username == name), None)
        table_data.append({
            "collaborator": name,
            "collaborations_2022_or_newer": count,
            "profile_post_count": post_count,
            "profile_category": profile_category
        })

collaborator_table_df = pd.DataFrame(table_data)
display(collaborator_table_df)
```
